```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datafortune = pd.read_csv('Fortune 500 2017.csv')
organised = datafortune
columns_to_delete = {'Website', 'Hqaddr', 'Hqtel', 'Ceo', 'Ceo-title', 'Address', 'Ticker', 'Fullname'}
```

```python
print(organised.Sector.nunique())
organised.Sector.unique()
```

```python
for col in organised.columns:
    print(col)
    print (organised[col].nunique(dropna = True))
```

```python
list_of_sectors = []
for sector in organised['Sector']:
    if not sector in list_of_sectors:
        list_of_sectors.append(sector)
print (list_of_sectors)
organised.isnull().sum()
```

```python
for sector in list_of_sectors:
    print('Doing sector ' + str(sector))
    avgemployees = 0
    avgrev = 0
    avgassets = 0
    avgtotshequity = 0
    count = 0
    index = 0
    for column in organised['Sector']:
        if column == sector:
            avgemployees += organised["Employees"][index]
            avgrev += organised['Revenues'][index]
            avgassets += organised['Assets'][index]
            avgtotshequity += organised['Totshequity'][index]
            count += 1
        index += 1
    avgemployees /= count
    avgrev /= count
    avgassets /= count
    avgtotshequity /= count
    print('the avg employees is ' + str(avgemployees))
    print('the avg rev is ' + str(avgrev))
    print('the avg assets is ' + str(avgassets))
    print('the avg totshequity is ' + str(avgtotshequity))
    print('The count is ' + str(count) + '\n')
```
